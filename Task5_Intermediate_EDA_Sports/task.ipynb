```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('matches.csv')
df1.head(10)
```

```python
team_list = df1['team1'].unique()
players = []

for team in team_list:

    rcb = df1[df1['winner'] == team]

    count = np.array(rcb['player_of_match'].value_counts()[:10])
    plt.bar(list(dict(rcb['player_of_match'].value_counts()[:10])),count)
    players.append(list(dict(rcb['player_of_match'].value_counts()[:1])))
    plt.title(team)
    plt.xlabel('Player of the match')
    plt.ylabel('Frequencies')
    plt.xticks(rotation='vertical')
    plt.savefig(team + '.png')
    plt.show()




```

# Best Players for the franchises

```python
best = pd.DataFrame([team_list,players])
best.head()
```
